## Question 6
What weather conditions are associated with flight delays leaving NYC? Use graphics to explore.

````python
weather_df = pd.read_csv('weather.csv')
weather_df


# group data by month
month_df = flights_df.groupby('month').mean()

# separate sum of departure delays
delay_by_month = month_df['dep_delay']  

# plot flight delays
plt.plot(delay_by_month, label = 'delay')
plt.ylabel('avgerage measure')
plt.xlabel('month')


# group weather data by month
we_month = weather_df.groupby('month').mean()

# splitting different variables 
wind_speed = we_month['wind_speed']
wind_gust = we_month['wind_gust']
precip = we_month['precip']
visib = we_month['visib']

# plot everything and create legend
plt.plot(wind_speed, label='wind speed')
plt.plot(wind_gust, label='wind gust')
plt.plot(precip, label='precipitation')
plt.plot(visib, label='visib')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
````

Wind speed and windgust seem to have an inverse effect on departure delays. When there is higher average wind speed/gust, there is a lower flight delay and vis versa. Visibility doesn't seem to have any effect on departure delays. Additionally, the dataset seems to be lacking information on precipitation.
